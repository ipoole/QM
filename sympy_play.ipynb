## Cartesian coordinates

```python
from sympy import *
from sympy.tensor.tensor import TensorIndexType, TensorIndex, TensorHead, tensor_indices
init_printing() 
```

```python
CartTIT = TensorIndexType('CartTIT', dummy_name='L')
i, j, k = tensor_indices('i j k', CartTIT)
CartVecTH = TensorHead('v', [CartTIT])
CartVecTH(i)
```

```python
CartVecTH(-i)
```

```python
cart_vec_repl = {CartTIT: diag(1,1), CartVecTH(i): [3,4]}
```

```python
expr = CartVecTH(i) * CartVecTH(-i)
expr, expr.replace_with_arrays(cart_vec_repl)
```

```python
CartTransTH = TensorHead('Lambda', [CartTIT, CartTIT])
CartTransTH(i, j)
```

```python
theta = symbols('theta')
rot_repl = {CartTIT: diag(1,1), CartTransTH(i, j): [[cos(theta), -sin(theta)], [sin(theta), cos(theta)]]}
```

```python
def rotation(angle, i0, i1):
    repl = {CartTransTH(i, j): [[cos(angle), -sin(angle)], [sin(angle), cos(angle)]]}
    expr = CartTransTH(i0, i1)
    return expr.replace_with_arrays(repl)
rotation(theta, i, j)
```

```python
rot_trans = CartTransTH(i, j)
rot_trans, rot_trans.replace_with_arrays(rot_repl)
```

```python
expr = CartTransTH(i, -j) * CartVecTH(j)
expr_inst = expr.replace_with_arrays({**rot_repl, **cart_vec_repl}).subs({theta:pi/2})
expr, expr_inst
```

```python
expr = CartTransTH(i, -j) * CartTransTH(j, -k) * CartVecTH(k)
expr_inst = expr.replace_with_arrays({**rot_repl, **cart_vec_repl}).subs({theta:pi/2})
expr, expr_inst
```

## 2D Spherical Surface coordinates
So this is a curved coordinate system, ($\theta, \phi$)

```python
from sympy import *
from sympy.tensor.tensor import TensorIndexType, TensorIndex, TensorHead, tensor_indices
init_printing() 
```

```python
r, theta = symbols('r theta')
```

```python
SphSurfTIT = TensorIndexType('SphSurfTIT', dummy_name='L')
i, j, k = tensor_indices('i j k', SphSurfTIT)
SphSurfTH = TensorHead('s', [SphSurfTIT])
SphSurfTH(i)
```

```python
# g_components = MutableDenseNDimArray([1, 0, 0, sp.sin(theta)**2], (2, 2))
g_components = [[1, 0], [0, sin(theta)**2]]
g_components
```

```python
repl = {SphSurfTIT: g_components, SphSurfTH(i):[0,pi/2]}
expr = SphSurfTH(i)
expr.replace_with_arrays(repl)
```

```python
repl = {SphSurfTIT: g_components, SphSurfTH(i):[0,pi/2]}
expr = SphSurfTH(i) * SphSurfTH(-i)
expr_inst = expr.replace_with_arrays(repl)
expr, expr_inst
```

```python
d_theta=pi/180
d_phi=pi/180
ds_expr = d_theta * d_phi * expr.replace_with_arrays(repl)
ds_expr
```
