```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=41
points = [(np.cos(theta), np.sin(theta)) for theta in np.linspace(0, 2*np.pi, n)]
circle = [(np.cos(theta), np.sin(theta)) for theta in np.linspace(0, 2*np.pi, 200)]
```

```python
step = 15

plt.plot(*zip(*circle))
for i in range(0,n):
    p1 = points[i%n]
    p2 = points[(i+step) % (n-1)]
    plt.plot(*zip(*[p1,p2]), color='r')


plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
